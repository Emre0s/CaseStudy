```python
agg_df = df.groupby(["COUNTRY" , "SOURCE" , "SEX" , "AGE"]).agg({"PRICE" : "mean"}).sort_values("PRICE" , ascending=False)
```

```python
agg_df
```

```python
agg_df.reset_index(inplace=True)
```

```python
agg_df
```

```python
araliklar = [0, 18, 23, 30, 40, 70]
agg_df['AGE'] = pd.cut(agg_df['AGE'], bins=araliklar)
```

```python
agg_df
```

```python
agg_df['AGE'] = agg_df['AGE'].astype(str)
```

```python
agg_df['customers_level_based'] = agg_df[["COUNTRY", "SOURCE", "SEX", "AGE"]].agg(lambda x: '_'.join(x).upper(), axis=1)
agg_df
```

```python
agg_df["SEGMENT"] = pd.qcut(agg_df["PRICE"], 4, labels=["D", "C", "B", "A"])
```

```python
agg_df
```

```python
agg_df.groupby("SEGMENT").agg({"PRICE" : ["size" , "sum" , "mean" , "max"]})
```
